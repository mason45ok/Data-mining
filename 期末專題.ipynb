```python
import re

def remove_mark(text):
    # 移除HTML標記
    text = re.sub(r'<[^>]+>', '', text)
    # text = text.replace('\n', '')
    text = text.replace(' ', '')
    text = text.replace('#', '')
    text = text.replace('*', '')
    text = text.replace('\n', '')
    return text
```

```python
import textwrap
from IPython.display import Markdown
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
import PyPDF2
import os
import glob


def extract_text_from_pdfs_in_directory(directory):
    text = ""
    pdf_files = glob.glob(os.path.join(directory, "./*.pdf"))
    for pdf_file in pdf_files:
        with open(pdf_file, "rb") as f:
            reader = PyPDF2.PdfReader(f)
            num_pages = len(reader.pages)
            for page_num in range(num_pages):
                page = reader.pages[page_num]
                text += page.extract_text()
    return text

# 使用上述函數從 PDF 中提取文字
pdf_text = extract_text_from_pdfs_in_directory("C:/Users/mason/Desktop/作業/科技系/教育資料探勘專題製作/Data-mining")
#For Debug
pdf_text
```

```python
def chatting(instruction):
  LAB_prompt = pdf_text + "這是我的國文教材"
  final_prompt = LAB_prompt + instruction
  model = genai.GenerativeModel('gemini-1.5-pro-latest')
  response = model.generate_content(final_prompt, stream=True)
  response.resolve()
  return to_markdown(response.text)
```

```python
import gradio as gr
import google.generativeai as genai

# 定義 chatting 函數
def chatting(lecture, topic, goal, duration, category):
    # 將使用者輸入的資訊組合成指示
    instruction = f"請幫我設計一堂關於'{lecture}'的課程，主題為'{topic}'，目標是'{goal}'，時長為'{duration}'小時，類別為'{category}'。"
    LAB_prompt = pdf_text + "，這是我的教材，" + instruction + "Use Traditional Chinese" + f"如果你認為主題和{lecture}無關，請直接回答與課程內容無關"
    final_prompt = LAB_prompt
    #print(lecture,topic,goal,duration,category)
    # 假設你已經有一個 GenerativeModel 的物件 model
    model = genai.GenerativeModel('gemini-1.5-pro-latest')
    response = model.generate_content(final_prompt, stream=True)
    response.resolve()
    return remove_mark(response.text)

# 定義 Gradio 介面
demo = gr.Interface(
    chatting,  # 使用 chatting 函數
    [
        gr.Dropdown(
            ["岳陽樓記/范仲淹", "鬼頭刀/廖鴻基","醉翁亭記/歐陽修","裨海紀遊選-北投硫穴記/郁永河","小城連作/鄭愁予","晚由六橋待月記/袁宏道","天才夢/張愛玲","前出師表/諸葛亮","一桿「稱仔」/賴和"], label="Lecture", info="想要哪一課的教案?"),
        gr.Textbox(label="課程主題", lines=3, value=""),
        gr.Textbox(label="教學目標", lines=3, value=""),
        gr.Number(label="課程時長(小時)", value=0),
        gr.Radio(["戶外參訪課程", "室內活動課程", "討論題目"], label="課程類別"),
    ],
    gr.Textbox(label="對話生成結果", lines=10),  # 顯示對話生成結果的文本框
)

if __name__ == "__main__":
    demo.launch()  # 啟動 Gradio 介面
```
